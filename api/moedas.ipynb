```python
import requests

cotacoes = requests.get('https://economia.awesomeapi.com.br/json/all').json()

print(cotacoes)
```

```python
print('Dolar: {}'.format(cotacoes['USD']['bid']))
print('Euro: {}'.format(cotacoes['EUR']['bid']))
print('BTC: {}'.format(cotacoes['BTC']['bid']))
```

```python
dolar_30_dias = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/30').json()
lista_cotacao_dolar = [float(item['bid']) for item in dolar_30_dias]
print(lista_cotacao_dolar)
```

```python
cotacoes_btc = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/365?start_date=20220101&end_date=20221231').json()
lista_cotacao_btc = [float(item['bid']) for item in cotacoes_btc]
lista_cotacao_btc.reverse()

print(lista_cotacao_btc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(lista_cotacao_btc)
plt.show()
```
